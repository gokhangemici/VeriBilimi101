```python
import numpy
import matplotlib.pyplot as plt

x = numpy.random.normal(5.0, 1.0, 1000) #mean = 5.0, standard_deviation = 1.0, 1000 random numbers come from normal data distribution
y = numpy.random.normal(10.0, 2.0, 1000) #mean = 10.0, standard_deviation = 2.0, 1000 numbers from normal distribution

plt.scatter(x, y)
plt.show()
```
