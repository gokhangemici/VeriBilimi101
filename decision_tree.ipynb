```python
import pandas as pd
from sklearn import tree
import pydotplus
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
import matplotlib.image as pltimg

df = pd.read_csv("C:\\Users\\ASUS\\OneDrive\\Masaüstü\\Veri bilimi 101\\shows.csv")
```

```python
df
```

```python
# change string values into numerical values
d = {'UK': 0, 'USA': 1, 'N': 2}
df['Nationality'] = df['Nationality'].map(d)
d = {'YES': 1, 'NO': 0}
df['Go'] = df['Go'].map(d)
```

```python
df
```

```python
features = ['Age', 'Experience', 'Rank', 'Nationality']

X = df[features]
y = df['Go']
```

```python
X
```

```python
y
```

```python

dtree = DecisionTreeClassifier()
dtree = dtree.fit(X, y)
print(dtree.predict([[40, 5, 9, 0]])) # 40 yasında 5 yıl deneyimli 9 puana sahip komedyenin şovuna gitmeli miyim
# bana çıktı olarak [1] git veya [0] gitme değerini döndürücek

print("[1] means 'GO'")
print("[0] means 'NO'")
```
